```python
import pandas as pd
df=pd.read_csv('sample_sales_data.csv')
df.head()
```

```python
df.info()
print(df.isnull().sum())
df=df.drop_duplicates()
df['date']=pd.to_datetime(df['date'])
df.columns=df.columns.str.lower()
df.head(50)
```

```python
print("Total sales: ",df['sales'].sum())
print("Average sales: ",df['sales'].mean())
print("Max sale: ",df['sales'].max())
print("Min sale: ",df['sales'].min())
print("unique products",df['product'].nunique())
```

```python
sales_by_region=df.groupby('region')['sales'].sum()
print(sales_by_region)
average_sales_product=df.groupby('product')['sales'].mean()
print(average_sales_product)
df['month']=df['date'].dt.month
monthly_sales=df.groupby('month')['sales'].sum()
print(monthly_sales)
```

```python
print(df[df['sales']>10000])
```

```python
print(df[(df['product']=='Laptop')& (df['quantity']>5)&(df['region']=='North')])
```

```python
print(df[df['region']=='North'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sales_by_region.plot(kind='bar',title='Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
monthly_sales.plot(kind='line',marker='o',title='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
sns.histplot(df['sales'],bins=20,kde=True,color='skyblue')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')
plt.show()
```

```python
top_products=df.groupby('product')['sales'].sum().nlargest(5)
print(top_products)
top_products.plot(kind='barh',title='Top 5 Products by Sales')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.show()
```
