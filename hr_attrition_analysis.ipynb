```python
import pandas as pd
df=pd.read_csv('hr_employee_attrition.csv')
df.head(10)
```

```python
df.info()
df.isnull().sum()
df=df.drop_duplicates()
df.columns=df.columns.str.lower()
df.head(50)
```

```python
print("Total Employees: ",df['employeeid'].count())
print("Total attrition: ",df[df['attrition']=='Yes'].shape[0])
total_employees=df['employeeid'].count()
total_attrition=df[df['attrition']=='Yes'].shape[0]
percent_of_attrition=(total_attrition/total_employees)*100
print("percentage of attrition: ",percent_of_attrition)
avg_age=df['age'].mean()
print("average age: ",avg_age)
print("Average salary: ",df['monthlyincome'].mean())
print("Average years at the company: ",df['yearsatcompany'].mean())
high_attr_dept=df[df['attrition']=='Yes'].groupby('department')['attrition'].count().sort_values(ascending=False)
print("Department with highest attrition: ",high_attr_dept.idxmax())
low_attr_job=df[df['attrition']=='No'].groupby('jobrole')['attrition'].count().sort_values(ascending=True)
print("Job role with lowest attrition: ",low_attr_job.idxmin())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
high_attr_dept.plot(kind='bar',title='Department with highest attrition')
plt.show()
```

```python
attr_gen = df.groupby('gender')['attrition'].count().sort_values(ascending=False)
attr_gen.plot(kind='pie', title='Attrition by gender')
plt.show()
```

```python
sns.histplot(df['monthlyincome'],bins=10,kde=True)
plt.title('Monthly income distribution')
plt.show()
```

```python
atrr_over=df.groupby('overtime')['attrition'].count().sort_values(ascending=False)
atrr_over.plot(kind='barh',title='Attrition by overtime')
plt.show()
```
