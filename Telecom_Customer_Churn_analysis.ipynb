```python
import pandas as pd
df=pd.read_csv('telecom_customer_churn.csv')
df.head(50)
```

```python
df.info()
df.isnull().sum()
df=df.drop_duplicates()
df.columns=df.columns.str.lower()
df.head(50)
```

```python
total_customers=df['customerid'].count()
print("Total Customers: ",total_customers)
churned_customers=df[df['churn']=='Yes'].count()['customerid']
print("Churned customers: ",churned_customers)
churn_rate=(churned_customers/total_customers)*100
print("Churn rate: ",churn_rate)
avg_tenure_churned=df[df['churn']=='Yes']['tenure'].mean()
print("Average tenure of churned customers: ",avg_tenure_churned)
avg_tenure_nonchurned=df[df['churn']=='No']['tenure'].mean()
print("Average tenure of non-churned customers: ",avg_tenure_nonchurned)
avg_monthcharge_churn=df[df['churn']=='Yes']['monthlycharges'].mean()
print("Average monthly charge of churned customers: ",avg_monthcharge_churn)
avg_monthcharge_nonchurn=df[df['churn']=='No']['monthlycharges'].mean()
print("Average monthly charge of non-churned customers: ",avg_monthcharge_nonchurn)
```

```python
churn_rate_contract_type=df.groupby('contracttype')['churn'].value_counts(normalize=True).unstack()
print("Churn rate by contract type: ")
print(churn_rate_contract_type)
import matplotlib.pyplot as plt
churn_rate_contract_type.plot(kind='bar',stacked=True)
plt.xlabel('Contract Type')
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Contract Type')
plt.show()
```

```python
churn_gender=df.groupby('gender')['churn'].value_counts().sort_values(ascending=False)
print("Churn rate by gender: ")
print(churn_gender)
churn_gender.plot(kind='pie')
plt.title('Churn Rate by Gender')
plt.show()
```
