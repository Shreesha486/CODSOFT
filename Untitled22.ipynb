```python
!pip install yfinance
import yfinance as yf
import pandas as pd
```

```python
apple=yf.download("AAPL",start="2020-01-01",end="2025-01-01")
apple.head()
```

```python
apple.reset_index(inplace=True)
apple.head()
```

```python
apple.columns=[col[0] if col[0]!='Price' else 'Date' for col in apple.columns.values]
apple.head()
```

```python
apple.info()
```

```python
missing_values = apple.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
duplicate_rows = apple.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
print("Highest closing price",round(apple['Close'].max(),3))
print("Average trading volume",round(apple['Volume'].mean()))
return_percent=round(((apple['Close'].iloc[-1])/(apple['Close'].iloc[0])*100))
print("Percentage of return: ",return_percent,"%")
```

```python
apple['Daily Return']=apple['Close'].pct_change()
print(apple['Daily Return'].describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(apple['Date'],apple['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Apple Stock Price Over Time')
plt.grid(True)
plt.show()
```

```python
apple['Daily Return'].hist(bins=50,figsize=(10,5))
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Returns')
plt.grid(True)
plt.show()
```

```python
stocks=yf.download(["AAPL","GOOG","MSFT"],start="2020-01-01",end="2025-01-01")
stocks.head()
```

```python
stocks.reset_index(inplace=True)
stocks.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(stocks['Date'],stocks['Close']['AAPL'],label='Apple')
plt.plot(stocks['Date'],stocks['Close']['GOOG'],label='Google')
plt.plot(stocks['Date'],stocks['Close']['MSFT'],label='Microsoft')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title(' Stock Price Over Time')
plt.grid(True)
plt.legend()
plt.show()
print(stocks.corr)
```
